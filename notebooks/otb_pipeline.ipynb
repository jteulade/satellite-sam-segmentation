```python


import os

# Set the OTB installation path
os.environ['PATH'] += os.pathsep + '/home/teulade/orfeo-toolbox/bin'

# Set the OTB application path
os.environ['OTB_APPLICATION_PATH'] = 'C:/path/to/otb/lib/otb/applications'

# Example of use with raster mode and mean-shift segmentation
import otbApplication

app = otbApplication.Registry.CreateApplication("Segmentation")

app.SetParameterString("in", "phr_xs.tif")
app.SetParameterString("mode","raster")
app.SetParameterString("mode.raster.out", "SegmentationRaster.tif")
app.SetParameterOutputImagePixelType("mode.raster.out", 3)
app.SetParameterString("filter","meanshift")

app.ExecuteAndWriteOutput()
```

```python
# Example of use with vector mode and watershed segmentation
import otbApplication
from osgeo import gdal

# Open the input bands
band2 = gdal.Open('B02.tif')
band3 = gdal.Open('B03.tif')
band4 = gdal.Open('B04.tif')
band8 = gdal.Open('B08.tif')

# Get the metadata of the first band
cols = band2.RasterXSize
rows = band2.RasterYSize
projection = band2.GetProjection()
geotransform = band2.GetGeoTransform()

# Create the output file
driver = gdal.GetDriverByName('GTiff')
output = driver.Create('output.tif', cols, rows, 4, gdal.GDT_Float32)

# Set the metadata
output.SetProjection(projection)
output.SetGeoTransform(geotransform)

# Write the bands to the output file
output.GetRasterBand(1).WriteArray(band2.GetRasterBand(1).ReadAsArray())
output.GetRasterBand(2).WriteArray(band3.GetRasterBand(1).ReadAsArray())
output.GetRasterBand(3).WriteArray(band4.GetRasterBand(1).ReadAsArray())
output.GetRasterBand(4).WriteArray(band8.GetRasterBand(1).ReadAsArray())

# Close the files
band2 = None
band3 = None
band4 = None
band8 = None
output = None

app = otbApplication.Registry.CreateApplication("Segmentation")

app.SetParameterString("in", "phr_xs.tif")
app.SetParameterString("mode","vector")
app.SetParameterString("mode.vector.out", "SegmentationVector.sqlite")
app.SetParameterFloat("filter.watershed.level", 0.2)
app.SetParameterFloat("filter.watershed.threshold", 0.1)
app.SetParameterFloat("mode.vector.minsize", 100)
app.SetParameterFloat("mode.vector.simplify float", 100)
app.SetParameterString("mode.vector.outmode", "ovw")
app.SetParameterString("filter","watershed")

app.ExecuteAndWriteOutput()
```
